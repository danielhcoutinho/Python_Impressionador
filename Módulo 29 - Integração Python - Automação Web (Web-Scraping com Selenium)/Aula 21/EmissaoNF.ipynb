```python
#importar a base de clientes
import pandas as pd

tabela = pd.read_excel("NotasEmitir.xlsx")
display(tabela.head())
```

```python
#para cada cliente - rodar o processo de emissao de nota fiscal
for linha in tabela.index:


    #preencher os dados da nota fiscal
    nome_razaosocial = navegador.find_element(By.NAME, 'nome').send_keys(tabela.loc[linha, "Cliente"])

    endereco = navegador.find_element(By.NAME, 'endereco').send_keys(tabela.loc[linha, "Endereço"])

    bairro = navegador.find_element(By.NAME, 'bairro').send_keys(tabela.loc[linha, "Bairro"])

    municipio = navegador.find_element(By.NAME, 'municipio').send_keys(tabela.loc[linha, "Municipio"])

    cep = navegador.find_element(By.NAME, 'cep').send_keys(str(tabela.loc[linha, "CEP"]))

    uf = navegador.find_element(By.NAME, 'uf').send_keys(tabela.loc[linha, "UF"])

    cnpj = navegador.find_element(By.NAME, 'cnpj').send_keys(str(tabela.loc[linha, "CPF/CNPJ"]))

    inscricao = navegador.find_element(By.NAME, 'inscricao').send_keys(str(tabela.loc[linha, "Inscricao Estadual"]))

    texto = tabela.loc[linha, "Descrição"]
    descricao = navegador.find_element(By.NAME, 'descricao').send_keys(texto)

    quantidade = navegador.find_element(By.NAME, 'quantidade').send_keys(str(tabela.loc[linha, "Quantidade"]))

    valor_unitario = navegador.find_element(By.NAME, 'valor_unitario').send_keys(str(tabela.loc[linha, "Valor Unitario"]))

    valor_total = navegador.find_element(By.NAME, 'total').send_keys(str(tabela.loc[linha, "Valor Total"]))

    #clicar em emitir nf

    botaoemitir = navegador.find_element(By.XPATH, '//*[@id="formulario"]/button').click()
```
