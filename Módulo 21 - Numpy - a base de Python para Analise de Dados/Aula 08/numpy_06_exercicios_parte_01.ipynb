## Exercício

Você é um analista de RH e tem os salários de todos os funcionários da sua empresa em um array NumPy. Seu trabalho é identificar quantos funcionários ganham acima da média. Use o seguinte array para sua análise: `salarios = np.array([3000, 2500, 3500, 4000, 2000, 4500, 3000, 3800, 4800])`.

```python
import numpy as np

salarios = np.array([3000, 2500, 3500, 4000, 2000, 4500, 3000, 3800, 4800])

media_salarial = np.mean(salarios)
print(media_salarial)

salarios_acima = np.where(salarios > media_salarial)


print(np.sum(salarios > media_salarial)) #printa a soma da qtd de salarios maior que a media
```

```python
print(np.count_nonzero(salarios > media_salarial)) #quantos valores nao sao 0

print(np.unique(salarios > media_salarial)) #quais valores unicos temos

print(np.unique(salarios > media_salarial, return_counts=True)) # conta quais valores unicos e retorna a contagem de quantos desse valor unico aparece

valores_unicos, contagem = np.unique(salarios > media_salarial, return_counts=True) # cria duas variaveis uma para os valores unicos e depois a contagem desses valores
print(valores_unicos)
print(contagem)
```
