### Read com pd.read_sql

```python
import pandas as pd
import sqlite3

conexao = sqlite3.connect("chinook.db") #cria a conexao com o banco de dados

tabela_clientes = pd.read_sql("SELECT *FROM customers", conexao) #faz a leitura do sql atraves do comando que selecionou tudo da tabela customers do banco de dados da conexao (2° parametro)

display(tabela_clientes[:3])

conexao.close()
```
